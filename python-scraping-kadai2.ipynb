```python
import requests
import pprint
from getpass import getpass

api_key=getpass('APIキーを入力してください。')

```

```python
search_station=input('駅名を入力してください。')
```

```python
url='https://maps.googleapis.com/maps/api/geocode/json'

response=requests.get(
    url,
    params={
        'key':api_key,
        'address':search_station
    }
)
```

```python
station_json=response.json()

pprint.pprint(station_json)
```

```python
for item in station_json['results']:
  lat=item['geometry']['location']['lat']
  longi=item['geometry']['location']['lng']
  print(lat)
  print(longi)
```

```python
url2='https://maps.googleapis.com/maps/api/place/nearbysearch/json'
response2=requests.get(
    url2,
    params={
        'location': f'{lat},{longi}',
        'language':'ja',
        'radius':500,
        'type':'restaurant',
        'key': api_key
    }
)

restrant=response2.json()

pprint.pprint(restrant)
```

```python
for item in restrant['results']:
  print(item['name'],item['vicinity'])
```
